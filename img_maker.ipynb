```python
import cv2
import pafy
```

```python
vidcap = cv2.VideoCapture('MMo_video/v1.mp4')
success,image = vidcap.read()
count = 0
frameRate = 30
while success :
    if count % frameRate == 0: 
        
        cv2.imwrite("./mmoRPG/frame%d.jpg" % (count//frameRate), image)     # save frame as JPEG file      
        success,image = vidcap.read()
        print('Add a new frame: id = ',count//frameRate, success)
    success,image = vidcap.read()
    count += 1
```

```python
def video_from_youtube(url):

#     url   = "https://www.youtube.com/watch?v=Xh_CqY9U3w4"
    video = pafy.new(url)
#     best  = video.streams()
    #documentation: https://pypi.org/project/pafy/

    capture = cv2.VideoCapture(video.streams[0].url)
    
    return capture
```

```python
import pafy
# url   = "https://www.youtube.com/watch?v=Xh_CqY9U3w4"
# video = pafy.new('Xh_CqY9U3w4',True,False,True,False)
video = pafy.new('Xh_CqY9U3w4')
best  = video.getbest(preftype="mp4")
vidcap = cv2.VideoCapture(best.url)
success,image = vidcap.read()
count = 0
frameRate = 30
while success :
    if count % frameRate == 0: 
        
        cv2.imwrite("./mmoRPG/v2_frame%d.jpg" % (count//frameRate), image)     # save frame as JPEG file      
        success,image = vidcap.read()
        print('Add a new frame: id = ',count//frameRate, success)
    success,image = vidcap.read()
    count += 1
```
